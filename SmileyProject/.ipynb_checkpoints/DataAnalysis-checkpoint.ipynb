## Predicting the next smiley
In order to optimise the inspection visits, it could be nice to predict the next smiley

We will try and predict the next smiley from the current smiley and industry sector, using a neural net.

First seperation into test and training data, and reformatting for the network

````python
def oneHotEncode(data, features):
    featdict = {feature:ix for ix, feature in enumerate(features)}
    output = np.zeros( (len(data), len(features)) )
    for d in enumerate(data):
        output[d[0], featdict[d[1]]] = 1
        
    return pd.DataFrame(output, columns=features, index=data.index)

def normalizeSmilies(x):
    return (x - ((4+1)/2)) / ((4-1)/2) # x - 
````

````python
training = pd.DataFrame()
validation = pd.Series()

counter = Counter(df['pixibranche'])

newDf = pd.DataFrame()

brancheCounter = {}

for b in counter.items():
    if (b[1] < 100):
        df = df[df.pixibranche != b[0]]
    else:
        brancheCounter[b[0]] = b[1]

counter1 = Counter(df['pixibranche'])

##Take 70% of each category for more even results - perhaps
for b in counter1.items():
    tmpDf = df[df.pixibranche == b[0]]

    tmpDf = tmpDf[['fjerdeseneste_kontrol','tredjeseneste_kontrol','naestseneste_kontrol','seneste_kontrol','pixibranche']]
    t = tmpDf.sample(frac = 0.7)
    v = tmpDf.loc[~tmpDf.index.isin(t.index)]
    training = pd.concat([t, training])
    validation = pd.concat([v, validation])
    
targets = training['pixibranche'].unique()

#Encode validation data
encoded_df1 = oneHotEncode(validation['pixibranche'], targets)
validation.drop('pixibranche', axis=1, inplace=True)
validation = pd.merge(validation, encoded_df1, left_index=True, right_index=True)


#Encode training data
encoded_df2 = oneHotEncode(training['pixibranche'], targets)
training.drop('pixibranche', axis=1, inplace=True)
training = pd.merge(training, encoded_df2, left_index=True, right_index=True)

tmpDict_T = {}

for nc in range(1,3):
    tmpDict_T[nc] = pd.concat([training.ix[:,nc-1:nc+1], training.ix[:,4:]], axis=1)
    tmpDict_T[nc].rename(columns={training.columns[nc]:'output', training.columns[nc-1]:'input'}, inplace=True)


tmpDict_T = pd.concat(tmpDict_T)
tmpDict_T = tmpDict_T.loc[(tmpDict_T.output.map(math.isnan)==False) & (tmpDict_T.input.map(math.isnan)==False)]

tmpDict_T = pd.concat([tmpDict_T.ix[1], tmpDict_T.ix[2]])

tmpDict_V = {}

for nc in range(1,3):
    tmpDict_V[nc] = pd.concat([validation.ix[:,nc-1:nc+1], validation.ix[:,4:]], axis=1)
    tmpDict_V[nc].rename(columns={validation.columns[nc]:'output', validation.columns[nc-1]:'input'}, inplace=True)


tmpDict_V = pd.concat(tmpDict_V)
tmpDict_V = tmpDict_V.loc[(tmpDict_V.output.map(math.isnan)==False) & (tmpDict_V.input.map(math.isnan)==False)]

tmpDict_V = pd.concat([tmpDict_V.ix[1], tmpDict_V.ix[2]])

print len(tmpDict_V)
````

````python
##Format data for training and test

y = tmpDict_T['output'].map(normalizeSmilies)

tmpDict_X = tmpDict_T.drop('output', axis=1)

tmpDict_X['input'] = tmpDict_X['input'].map(normalizeSmilies)

tmpDict_X[targets] = tmpDict_X[targets].applymap(lambda x: -1 if x == 0 else 1)

X = tmpDict_X.as_matrix()
y = y.as_matrix().reshape(len(y),1)


y_v = tmpDict_V['output'].map(normalizeSmilies)

tmpDict_X_V = tmpDict_V.drop('output', axis=1)

tmpDict_X_V['input'] = tmpDict_X_V['input'].map(normalizeSmilies)

tmpDict_X_V[targets] = tmpDict_X_V[targets].applymap(lambda x: -1 if x == 0 else 1)

X_validation = tmpDict_X_V.as_matrix()
y_validation = y_v.as_matrix().reshape(len(y_v),1)
````

````python
## Training
# sigmoid function
def nonlin(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))


# seed random numbers to make calculation
# deterministic (just a good practice)
np.random.seed(1203)


# initialize weights randomly with mean 0
syn0 = 2*np.random.random((X.shape[1],1)) - 1

error = []

for x in range(10):
    stepSize = 1
    currentStep = stepSize
    
    for iter in range(0,X.shape[0], stepSize):
    
        if(currentStep > X.shape[0]):
            currentStep = X.shape[0] - 1

        currentRange = range(iter, currentStep)

        # forward propagation
        l0 = X[np.array(currentRange)]
        l1 = nonlin(np.dot(l0,syn0))
        
        # how much did we miss?
        l1_error = y[iter:currentStep] - l1

        # multiply how much we missed by the 
        # slope of the sigmoid at the values in l1
        l1_delta = l1_error * nonlin(l1,True)

        # update weights
        syn0 += np.dot(l0.T,l1_delta)

        currentStep = currentStep + stepSize
    
        error.append(l1_error[0,0])
````

````python
plt.plot(error[0:5000:10])
plt.xlabel('Epoch number')
plt.ylabel('Train error')
plt.grid(True)
````

The neural net fast learns to guess on a happy smiley as this is correct most of the time, no matter what sector or smiley you received last time. Thus the network is not able to properly predict the smiley given only this data as input. Similar results were found using linear regression. In order to compensate for this problem, a different and more balanced featureset must be used to get any results. The problem with getting such features is that they are not readily available, especially not since the dataset covers many business types. For an isolated problem only focused on restaurants a feature such as ratings might be useable.

````python
##Validaiton/Test
correntCount = 0
errorCount = 0

for iter in range(0,X_validation.shape[0]):
    currentRange = range(iter, iter+1)
    
    l0 = X[np.array(currentRange)]
    l1 = nonlin(np.dot(l0,syn0))
    if(l1[0,0] == y_validation[currentRange]):
        correntCount = correntCount + 1
    else:
        errorCount = errorCount + 1
    
    print "Value: ", y_validation[currentRange]
    print "Prediction: ", l1[0,0]
````

## Location of a bad smiley

In order to judge the location of smileys, and bad smileys in particular the data has to be normalised based on population.
As the post code of each production unit is given in the dataset, it is possible to cross reference it with other post code data.

Data for population density, is not readily available, but map data and population data are available on the internet. These can be combined to calculate a population density estimate for each post code.

As a bonus the map data can be used for visualization using d3 and the geojson format.

### Area library
To calculate the area of a geojson feature the following library is used

````python
# Area of geojson polygon
from math import pi, sin

WGS84_RADIUS = 6378137

def rad(value):
    return value * pi / 180

def ring__area(coordinates):
    """
    Calculate the approximate _area of the polygon were it projected onto
        the earth.  Note that this _area will be positive if ring is oriented
        clockwise, otherwise it will be negative.
    Reference:
        Robert. G. Chamberlain and William H. Duquette, "Some Algorithms for
        Polygons on a Sphere", JPL Publication 07-03, Jet Propulsion
        Laboratory, Pasadena, CA, June 2007 http://trs-new.jpl.nasa.gov/dspace/handle/2014/40409
    @Returns
    {float} The approximate signed geodesic _area of the polygon in square meters.
    """

    assert isinstance(coordinates, list)

    _area = 0
    coordinates_length = len(coordinates)

    if coordinates_length > 2:
        for i in range(0, coordinates_length):
            if i == (coordinates_length - 2):
                lower_index = coordinates_length - 2
                middle_index = coordinates_length - 1
                upper_index = 0
            elif i == (coordinates_length - 1):
                lower_index = coordinates_length - 1
                middle_index = 0
                upper_index = 1
            else:
                lower_index = i
                middle_index = i + 1
                upper_index = i + 2

            p1 = coordinates[lower_index]
            p2 = coordinates[middle_index]
            p3 = coordinates[upper_index]

            _area += (rad(p3[0]) - rad(p1[0])) * sin(rad(p2[1]))

        _area = _area * WGS84_RADIUS * WGS84_RADIUS / 2

    return _area

def polygon__area(coordinates):

    assert isinstance(coordinates, list)

    _area = 0
    if len(coordinates) > 0:
        _area += abs(ring__area(coordinates[0]))

        for i in range(1, len(coordinates)):
            _area -= abs(ring__area(coordinates[i]))

    return _area


def area(geometry):

    if isinstance(geometry, str):
        geometry = json.loads(geometry)

    assert isinstance(geometry, dict)

    _area = 0

    if geometry['type'] == 'Polygon':
        return polygon__area(geometry['coordinates'])
    elif geometry['type'] == 'MultiPolygon':
        for i in range(0, len(geometry['coordinates'])):
            _area += polygon__area(geometry['coordinates'][i])

    elif geometry['type'] == 'GeometryCollection':
        for i in range(0, len(geometry['geometries'])):
            _area += area(geometry['geometries'][i])

    return _area
````

### Cross-referencing postcode, population, and smiley datasets
For each postcode a set of features is calculated and packed for d3 visualization
if the map feature covers multiple postcodes, the features are summed across them all.

The features are:
* Population
* Area
* Postcode name
* Number of production units
* Average Smiley rating among production units in postcode

which can be used to calculate the following:
* Population density
* Production units per population density

These are dumped to a common geojson file for d3 to load and visualise.

````python
# Load data
with open('../data/postnumre2008-07-01.geojson') as data_file:    
    mapdata = json.load(data_file)

befolkning = pd.read_csv('../data/PNR1_2005_befolkning.csv')

# Extract postnummer
befolkning['Postnummer'] = [int(s[0:4]) for s in befolkning.Navn]

# Match postnummer and extend geojson data
for feature in mapdata['features']:
    # print progress
    print feature['properties']['Description'].split('<td>')[1].split('</td>')[0]
    
    # Extract groups of postnummer and create masks
    pgroup = feature['properties']['Name'].split(' - ')
    if (len(pgroup) > 1):
        mask = (int(pgroup[0]) <= befolkning.Postnummer) & (befolkning.Postnummer <= int(pgroup[1]))
        dfmask = (int(pgroup[0]) <= df.postnr) & (df.postnr <= int(pgroup[1]))
    else:
        mask = befolkning.Postnummer == int(pgroup[0])
        dfmask = df.postnr == int(pgroup[0])
    
    # clean and extend geojson
    feature['properties']['Indbyggere']  = sum(befolkning.loc[mask].Indbyggere)
    feature['properties']['Description'] = feature['properties']['Description'].split('<td>')[1].split('</td>')[0]
    feature['properties']['Areal']       = area(feature['geometry'])/1e6
    feature['properties']['SmileyVirksomheder'] = sum(dfmask)
    feature['properties']['MostCommonPixi'] = Counter(df.loc[dfmask].pixibranche).most_common()[0][0]
    feature['properties']['AverageSmiley'] = np.mean(rating.loc[dfmask])

with open('../data/postnumre_clean.geojson', 'w') as fp:
    json.dump(mapdata, fp)
````

## Timing of smiley inspections

The different sectors are inspected using different interval schemes based on the current rating, and random checks. Is it possible for us to group these?

- - -
First all the inspections are seperated into a new dataset. This dataset contains location, postcode, sector, rating, and time until the next inspection, calculated in older inspections.

````python
# extract kontroller into one column
columns = ['','naest', 'tredje', 'fjerde']
kontroller = {};
for nc in range(0,len(columns)):
    # generate names
    column = columns[nc] + 'seneste_kontrol'
    nextColumn = columns[nc-1] + 'seneste_kontrol' if nc-1 >= 0 else None;
    
    # generate selector
    selector = [column, column + '_dato', 'postnr', 'Geo_Lat', 'Geo_Lng', 'pixibranche','Elite_Smiley'];
    if nextColumn:
        selector.extend([nextColumn+ '_dato'])
    
    # select data
    kontroller[column] = df[selector]

    #rename
    kontroller[column].rename(columns={column: 'kontrol', column + '_dato': 'dato'}, inplace=True)
    if nextColumn:
        kontroller[column].rename(columns={nextColumn + '_dato': 'next_dato'}, inplace=True)

# concatenate dict
kontroller = pd.concat(kontroller)

# remove nans
kontroller = kontroller.loc[kontroller.kontrol.map(math.isnan) == False]

branchelabels = {label: i for i, label in enumerate(set(kontroller.pixibranche))}
kontroller['pixibranche_id'] = kontroller.pixibranche.map(lambda b: branchelabels[b])

kontroller['deltat'] = pd.to_datetime(kontroller.next_dato)-pd.to_datetime(kontroller.dato)
kontroller['deltat'] = kontroller['deltat'].map(lambda x: x / np.timedelta64(1,'D'));
````

by looking at a scatter plot with high alpha per data point, a light histogram is made for all sectors
This first inspection of the dataset shows different inspection patterns, many with peaks around three months, one, and two years.

````python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.scatter(kontroller.deltat, kontroller.pixibranche_id, alpha=0.01, edgecolor='none');
plt.ylabel('Branche id')
plt.xlim(0,1000);
plt.xlabel("Time (Days)")
plt.ylim(-1,48);
plt.grid(True)
plt.title('Distribution of time between inspections for the different sectors');
````

Next, histogram data is calculated for each sector and for each current smiley rating and dumped to a file for visualization.

````python

nanmask = kontroller.deltat.map(math.isnan) == False
bins = range(0,1000,4*7)
histdata = {'bins':bins}
histdata['items'] = []
for branche in branchelabels:
    b_id = branchelabels[branche]
    
    buf = {}
    buf['deltatHist'] = {}
    bdata = kontroller.loc[nanmask & (kontroller.pixibranche_id == b_id)]
    buf['antal'] = len(bdata)
    for i in range(0,5):
        if i == 0: # elite
            d = np.histogram(bdata.loc[(bdata.Elite_Smiley == 1)].deltat.map(round),bins)[0]
        else:
            d = np.histogram(bdata.loc[(bdata.kontrol == i) & ((bdata.Elite_Smiley != 1))].deltat.map(round),bins)[0]
        
        buf['deltatHist'][i] = d.tolist()
    buf['branche'] = branche;
    
    histdata['items'].append(buf);

with open('../data/kontroltid_hist_data.json', 'w') as fp:
    json.dump(histdata, fp)
````

Grouping is done with K-Means in matlab as the scikit learn K-Means implementation dows not include the possibility of a custum distance measure.

The wanted distance measure is the cosine similarity, which is good at detecting shapes in vector data as it ignores the scaling.
The cosine similarity is a measure of angular distance in the n-dimensional vector space that is the training data.

The following Matlab code is run to cluster on the timing of the **happy but not elite** smiley inspections. This is chosen as the shape of this group seems to be quite characteristic, and because not all sectors can attain elite status.
